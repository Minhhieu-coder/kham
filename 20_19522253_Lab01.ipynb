```python
# Đọc cả hai file CSV
import pandas as pd
responses = pd.read_csv('kaggle_survey_2021_responses.csv')
metadata = pd.read_csv('metadata.csv')

# Xem kích thước và các cột của từng file
def quick_overview(df, name):
    print(f'--- {name} ---')
    print('Shape:', df.shape)
    print('Columns:', df.columns.tolist())
    display(df.head(3))
    print('\n')

quick_overview(responses, 'responses')
quick_overview(metadata, 'metadata')
```
