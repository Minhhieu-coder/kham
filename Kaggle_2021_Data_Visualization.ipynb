## 1. üì• T·∫£i v√† Kh√°m ph√° D·ªØ li·ªáu

```python
# ƒê·ªçc d·ªØ li·ªáu
df = pd.read_csv('kaggle_survey_2021_responses.csv', low_memory=False)

# L∆∞u h√†ng ƒë·∫ßu ti√™n (m√¥ t·∫£ c√¢u h·ªèi) v√† lo·∫°i b·ªè
questions = df.iloc[0].copy()
df = df.iloc[1:].reset_index(drop=True)

print(f'üìä K√≠ch th∆∞·ªõc d·ªØ li·ªáu: {df.shape[0]} ng∆∞·ªùi tham gia, {df.shape[1]} c·ªôt')
print(f'\nüìã C√°c c·ªôt ch√≠nh:')
print(df.columns.tolist()[:20])
```

```python
# Xem m·∫´u d·ªØ li·ªáu
df.head(10)
```

```python
# Th·ªëng k√™ missing values
missing_stats = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing_stats / len(df) * 100).round(2)

fig, ax = plt.subplots(figsize=(14, 6))
top_missing = missing_percent.head(30)
colors = plt.cm.Reds(np.linspace(0.3, 0.9, len(top_missing)))
bars = ax.barh(range(len(top_missing)), top_missing.values, color=colors)
ax.set_yticks(range(len(top_missing)))
ax.set_yticklabels(top_missing.index, fontsize=8)
ax.set_xlabel('T·ª∑ l·ªá Missing (%)')
ax.set_title('üîç Top 30 C·ªôt c√≥ Missing Values Cao Nh·∫•t', fontsize=14, fontweight='bold')
ax.invert_yaxis()
for i, (bar, val) in enumerate(zip(bars, top_missing.values)):
    ax.text(val + 0.5, i, f'{val:.1f}%', va='center', fontsize=8)
plt.tight_layout()
plt.show()
```

## 2. üë• Ph√¢n t√≠ch Nh√¢n kh·∫©u h·ªçc

```python
# 2.1 Ph√¢n b·ªë Tu·ªïi
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Pie chart
age_counts = df['Q1'].value_counts()
colors = plt.cm.Set3(np.linspace(0, 1, len(age_counts)))
explode = [0.02] * len(age_counts)
axes[0].pie(age_counts.values, labels=age_counts.index, autopct='%1.1f%%', 
            colors=colors, explode=explode, shadow=True, startangle=90)
axes[0].set_title('üéÇ Ph√¢n b·ªë Tu·ªïi (Pie Chart)', fontsize=14, fontweight='bold')

# Bar chart
age_order = ['18-21', '22-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-69', '70+']
age_sorted = df['Q1'].value_counts().reindex(age_order).dropna()
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(age_sorted)))
bars = axes[1].bar(age_sorted.index, age_sorted.values, color=colors, edgecolor='black', linewidth=0.5)
axes[1].set_xlabel('Nh√≥m tu·ªïi')
axes[1].set_ylabel('S·ªë l∆∞·ª£ng')
axes[1].set_title('üéÇ Ph√¢n b·ªë Tu·ªïi (Bar Chart)', fontsize=14, fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 2.2 Ph√¢n b·ªë Gi·ªõi t√≠nh
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

gender_counts = df['Q2'].value_counts()
colors = ['#3498db', '#e74c3c', '#2ecc71', '#9b59b6', '#f39c12']

# Donut chart
wedges, texts, autotexts = axes[0].pie(gender_counts.values, labels=gender_counts.index,
                                        autopct='%1.1f%%', colors=colors[:len(gender_counts)],
                                        wedgeprops=dict(width=0.6), startangle=90)
axes[0].set_title('üë´ Ph√¢n b·ªë Gi·ªõi t√≠nh (Donut Chart)', fontsize=14, fontweight='bold')

# Horizontal bar
y_pos = np.arange(len(gender_counts))
axes[1].barh(y_pos, gender_counts.values, color=colors[:len(gender_counts)], edgecolor='black')
axes[1].set_yticks(y_pos)
axes[1].set_yticklabels(gender_counts.index)
axes[1].set_xlabel('S·ªë l∆∞·ª£ng')
axes[1].set_title('üë´ Ph√¢n b·ªë Gi·ªõi t√≠nh (Horizontal Bar)', fontsize=14, fontweight='bold')
for i, v in enumerate(gender_counts.values):
    axes[1].text(v + 100, i, f'{v} ({v/len(df)*100:.1f}%)', va='center')

plt.tight_layout()
plt.show()
```

```python
# 2.3 Ph√¢n b·ªë theo Qu·ªëc gia - Top 20
fig, ax = plt.subplots(figsize=(14, 8))

country_counts = df['Q3'].value_counts().head(20)
colors = plt.cm.coolwarm(np.linspace(0.2, 0.8, len(country_counts)))

bars = ax.barh(range(len(country_counts)), country_counts.values, color=colors)
ax.set_yticks(range(len(country_counts)))
ax.set_yticklabels(country_counts.index)
ax.set_xlabel('S·ªë l∆∞·ª£ng ng∆∞·ªùi tham gia')
ax.set_title('üåç Top 20 Qu·ªëc gia c√≥ nhi·ªÅu ng∆∞·ªùi tham gia nh·∫•t', fontsize=14, fontweight='bold')
ax.invert_yaxis()

for i, (bar, val) in enumerate(zip(bars, country_counts.values)):
    ax.text(val + 50, i, f'{val} ({val/len(df)*100:.1f}%)', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 2.4 Tr√¨nh ƒë·ªô h·ªçc v·∫•n
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

edu_counts = df['Q4'].value_counts()
colors = plt.cm.Spectral(np.linspace(0, 1, len(edu_counts)))

# Pie chart
axes[0].pie(edu_counts.values, labels=None, autopct='%1.1f%%', colors=colors, startangle=90)
axes[0].legend(edu_counts.index, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
axes[0].set_title('üéì Tr√¨nh ƒë·ªô H·ªçc v·∫•n', fontsize=14, fontweight='bold')

# Bar chart
bars = axes[1].bar(range(len(edu_counts)), edu_counts.values, color=colors, edgecolor='black')
axes[1].set_xticks(range(len(edu_counts)))
axes[1].set_xticklabels([x[:20]+'...' if len(x) > 20 else x for x in edu_counts.index], rotation=45, ha='right', fontsize=8)
axes[1].set_ylabel('S·ªë l∆∞·ª£ng')
axes[1].set_title('üéì Tr√¨nh ƒë·ªô H·ªçc v·∫•n (Bar Chart)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 2.5 Vai tr√≤ c√¥ng vi·ªác
fig, ax = plt.subplots(figsize=(12, 8))

role_counts = df['Q5'].value_counts()
colors = plt.cm.tab20(np.linspace(0, 1, len(role_counts)))

bars = ax.barh(range(len(role_counts)), role_counts.values, color=colors)
ax.set_yticks(range(len(role_counts)))
ax.set_yticklabels(role_counts.index, fontsize=9)
ax.set_xlabel('S·ªë l∆∞·ª£ng')
ax.set_title('üíº Ph√¢n b·ªë Vai tr√≤ C√¥ng vi·ªác', fontsize=14, fontweight='bold')
ax.invert_yaxis()

for i, (bar, val) in enumerate(zip(bars, role_counts.values)):
    ax.text(val + 50, i, f'{val}', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## 3. üíª Ph√¢n t√≠ch Ng√¥n ng·ªØ L·∫≠p tr√¨nh

```python
# 3.1 Ng√¥n ng·ªØ l·∫≠p tr√¨nh ƒë∆∞·ª£c s·ª≠ d·ª•ng
prog_cols = [col for col in df.columns if col.startswith('Q7_Part') and 'OTHER' not in col]

prog_lang_counts = {}
for col in prog_cols:
    lang = df[col].dropna().unique()
    if len(lang) > 0:
        lang_name = lang[0] if isinstance(lang[0], str) else 'Unknown'
        prog_lang_counts[lang_name] = df[col].notna().sum()

prog_df = pd.DataFrame(list(prog_lang_counts.items()), columns=['Language', 'Count'])
prog_df = prog_df.sort_values('Count', ascending=True)

fig, ax = plt.subplots(figsize=(12, 8))
colors = plt.cm.RdYlGn(np.linspace(0.2, 0.8, len(prog_df)))
bars = ax.barh(prog_df['Language'], prog_df['Count'], color=colors)
ax.set_xlabel('S·ªë ng∆∞·ªùi s·ª≠ d·ª•ng')
ax.set_title('üíª Ng√¥n ng·ªØ L·∫≠p tr√¨nh Ph·ªï bi·∫øn', fontsize=14, fontweight='bold')

for bar in bars:
    width = bar.get_width()
    ax.text(width + 100, bar.get_y() + bar.get_height()/2, f'{int(width)} ({width/len(df)*100:.1f}%)',
            va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 3.2 Ng√¥n ng·ªØ ƒë∆∞·ª£c khuy·∫øn ngh·ªã h·ªçc ƒë·∫ßu ti√™n
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

recommend_counts = df['Q8'].value_counts()
colors = plt.cm.Set2(np.linspace(0, 1, len(recommend_counts)))

# Pie chart
axes[0].pie(recommend_counts.values[:8], labels=recommend_counts.index[:8], autopct='%1.1f%%',
           colors=colors, startangle=90, explode=[0.05]*min(8, len(recommend_counts)))
axes[0].set_title('üìö Ng√¥n ng·ªØ Khuy·∫øn ngh·ªã H·ªçc ƒê·∫ßu ti√™n', fontsize=12, fontweight='bold')

# Bar chart
bars = axes[1].bar(recommend_counts.index[:10], recommend_counts.values[:10], color=colors[:10], edgecolor='black')
axes[1].set_xlabel('Ng√¥n ng·ªØ')
axes[1].set_ylabel('S·ªë ng∆∞·ªùi khuy·∫øn ngh·ªã')
axes[1].set_title('üìö Ng√¥n ng·ªØ Khuy·∫øn ngh·ªã (Bar)', fontsize=12, fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 4. üõ†Ô∏è Ph√¢n t√≠ch IDE v√† C√¥ng c·ª•

```python
# 4.1 IDE ƒë∆∞·ª£c s·ª≠ d·ª•ng
ide_cols = [col for col in df.columns if col.startswith('Q9_Part') and 'OTHER' not in col]

ide_counts = {}
for col in ide_cols:
    ide_vals = df[col].dropna().unique()
    if len(ide_vals) > 0:
        ide_name = str(ide_vals[0]).strip()
        ide_counts[ide_name] = df[col].notna().sum()

ide_df = pd.DataFrame(list(ide_counts.items()), columns=['IDE', 'Count'])
ide_df = ide_df.sort_values('Count', ascending=True)

fig, ax = plt.subplots(figsize=(12, 8))
colors = plt.cm.plasma(np.linspace(0.2, 0.8, len(ide_df)))
bars = ax.barh(ide_df['IDE'], ide_df['Count'], color=colors)
ax.set_xlabel('S·ªë ng∆∞·ªùi s·ª≠ d·ª•ng')
ax.set_title('üõ†Ô∏è IDE v√† Notebooks Ph·ªï bi·∫øn', fontsize=14, fontweight='bold')

for bar in bars:
    width = bar.get_width()
    ax.text(width + 100, bar.get_y() + bar.get_height()/2, f'{int(width)}',
            va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## 5. ü§ñ Ph√¢n t√≠ch ML Frameworks

## 8. üìà Ph√¢n t√≠ch T∆∞∆°ng quan v√† Insights

```python
# 8.1 Tu·ªïi vs Gi·ªõi t√≠nh
fig, ax = plt.subplots(figsize=(12, 6))

age_gender = pd.crosstab(df['Q1'], df['Q2'])
age_order = ['18-21', '22-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-69', '70+']
age_gender = age_gender.reindex(age_order).dropna()

age_gender.plot(kind='bar', stacked=True, ax=ax, colormap='Set2', edgecolor='black')
ax.set_xlabel('Nh√≥m tu·ªïi')
ax.set_ylabel('S·ªë l∆∞·ª£ng')
ax.set_title('üë• Ph√¢n b·ªë Tu·ªïi theo Gi·ªõi t√≠nh (Stacked Bar)', fontsize=14, fontweight='bold')
ax.tick_params(axis='x', rotation=45)
ax.legend(title='Gi·ªõi t√≠nh', bbox_to_anchor=(1.02, 1))

plt.tight_layout()
plt.show()
```

```python
# 8.2 Vai tr√≤ c√¥ng vi·ªác vs H·ªçc v·∫•n (Heatmap)
fig, ax = plt.subplots(figsize=(14, 10))

role_edu = pd.crosstab(df['Q5'], df['Q4'])
role_edu_norm = role_edu.div(role_edu.sum(axis=1), axis=0) * 100

sns.heatmap(role_edu_norm, annot=True, fmt='.1f', cmap='YlOrRd', ax=ax, linewidths=0.5)
ax.set_xlabel('Tr√¨nh ƒë·ªô H·ªçc v·∫•n')
ax.set_ylabel('Vai tr√≤ C√¥ng vi·ªác')
ax.set_title('üî• Heatmap: Vai tr√≤ C√¥ng vi·ªác vs H·ªçc v·∫•n (%)', fontsize=14, fontweight='bold')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
